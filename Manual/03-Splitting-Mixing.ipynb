# Imports

```python
import LightPipes as lp
import numpy as np
import matplotlib.pyplot as plt
```

# Young's Experiment

```python
# Parameters
size = 20.0 * lp.mm
wave = 5 * lp.um
N = 512

z = 50 * lp.cm
R = 0.15 * lp.mm
d = 1.2 * lp.mm

# Generate and propagate the Field
Field  = lp.Begin(size, wave, N)
Field1 = lp.CircAperture(R, -d/2 , 0, Field)
Field2 = lp.CircAperture(R, d/2, 0, Field)    
Field = lp.BeamMix(Field1,Field2)
Field = lp.Fresnel(z,Field)
I = lp.Intensity(Field, flag=2)

# Draw the Field in 2D
I = lp.Intensity(Field, flag=0)

plt.imshow(I, extent=(0, size / lp.mm, 0, size / lp.mm), cmap="rainbow")
plt.colorbar()
plt.title("Field")
plt.xlabel('Position (mm)')
plt.ylabel('Position (mm)')
plt.show()
```
