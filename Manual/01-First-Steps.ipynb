# Apertures and Screens
We can start by modeling various apertures and screens! The x and y parameters are the offset from the center, where x points left to right and y points up to down.

```python
# Grid size
size = 10 * lp.mm

# Wavelength of the light used
wave = 500 * lp.nm

# Grid dimension (we can make it go up to about 2000 with no problems)
N = 256

# Create the field
Field = lp.Begin(size, wave, N)

# Create a circular aperture
R = 2 * lp.mm
xc = 1 * lp.mm
yc = 1 * lp.mm

Field = lp.CircAperture(R, xc, yc, Field)
I = lp.Intensity(Field, flag=0)

plt.imshow(I, extent=(0, size / lp.mm, 0, size / lp.mm))
plt.colorbar()
plt.title("Field")
plt.xlabel('Position (mm)')
plt.ylabel('Position (mm)')
plt.show()
```

We can also make interesing combinations of various apertures.

```python
# Grid size
size = 10 * lp.mm

# Wavelength of the light used
wave = 500 * lp.nm

# Grid dimension (we can make it go up to about 2000 with no problems)
N = 256

# Create the field
Field = lp.Begin(size, wave, N)

# Create a circular screen
R = 1 * lp.mm
xc = 3 * lp.mm
yc = 3 * lp.mm

Field = lp.CircScreen(R, xc, yc, Field)

# Create a diamond screen
wx = 2 * lp.mm
wy = 2 * lp.mm
xs = 1 * lp.mm
ys = 1 * lp.mm
phis = 45

Field = lp.RectScreen(wx, wy, xs, ys, phis, Field)

# Draw the Field
I = lp.Intensity(Field, flag=0)

plt.imshow(I, extent=(0, size / lp.mm, 0, size / lp.mm))
plt.colorbar()
plt.title("Field")
plt.xlabel('Position (mm)')
plt.ylabel('Position (mm)')
plt.show()
```

## Small Apertures

Small apertures have to take into consederation the finite resolution of our Field array.

```python
# Grid size
size = 8 * lp.mm

# Wavelength of the light used
wave = 500 * lp.nm

# Grid dimension (we can make it go up to about 2000 with no problems)
N = 8

# Space between pixels
h = size / N
print(h)

# Create the field
Field = lp.Begin(size, wave, N)
print(Field.field)

# Create a slit
Field = lp.RectScreen(Field, 4*h, h)
print(Field.field)
print(Field.field[N//2])
print((N-sum(Field.field[N//2])) * h)

# Draw the Field
I = lp.Intensity(Field, flag=0)

plt.imshow(I, extent=(-(size/lp.mm)//2, (size/lp.mm)//2, -(size/lp.mm)//2, (size/lp.mm)//2))
plt.colorbar()
plt.title("Field")
plt.xlabel('Position (mm)')
plt.ylabel('Position (mm)')
plt.show()
```

# Graphing and Visualizations

We can interpolate a lower resolution grid to obtain a higher resolution result faster!

```python
# Grid size
size = 10 * lp.mm

# Wavelength of the light used
wave = 500 * lp.nm

# Grid dimension (we can make it go up to about 2000 with no problems)
N = 256

# Create the field
Field = lp.Begin(size, wave, N)

# Create a circular screen
R = 2 * lp.mm
xc = 0 * lp.mm
yc = 0 * lp.mm
T = 0.8

Field = lp.GaussAperture(R, xc, yc, Field)

# Create a smaller grid
newN = N // 4

# Interpolate the smaller Field
Field = lp.Interpol(Field, size, newN)

I = lp.Intensity(0, Field)
I = np.array(I)

# Draw the Field in 2D
I = lp.Intensity(Field, flag=0)

plt.imshow(I, extent=(0, size / lp.mm, 0, size / lp.mm))
plt.colorbar()
plt.title("Field")
plt.xlabel('Position (mm)')
plt.ylabel('Position (mm)')
plt.show()

# Draw the field in 3D
xx = np.arange(newN)
yy = np.arange(newN)

xx, yy = np.meshgrid(xx, yy)

fig = plt.figure(figsize=(10, 6))
ax = plt.axes(projection="3d")
ax.plot_surface(xx, yy, I, rstride = 1, cstride = 1, cmap = "rainbow", linewidth = 0)

ax.set_label("Intensity")

plt.show()
```
